```python
import numpy as np
x=np.linspace(0,10,100)
"""축을 만드는 함수 linspace"""

fig=plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
"""show()를 하지 않아도 jupyter notebook에서는 IPython display를 통해 자동적으로 보여준다. 파이썬에서 print()안 해도 나오는 것도 마찬가지이다."""
```

```python
def my_plot():
    fig=plt.figure()   """figure 만들기, 흰 바탕"""
    x=np.linspace(0,10,11)  
    y=x**2

    fig.subplots().plot(x, y)
    fig.show()

my_plot()
```

```python
"""매트랩 스타일의 인터페이스"""

"""플롯 그림 생성"""
plt.figure()
"""두개의 패널 중 첫번째 패널을 생성하고 현재 축을 설정"""
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

"""두번째 패널을 생성하고 현재축을 설정"""
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
"""객체지향 인터페이스"""
"""먼저 플롯 그리드를 생성"""
"""ax는 두개의 축 객체의 배열이 됨"""
fig, ax= plt.subplots(2)

"""적절한 객체에서 plot()메서드를 호출"""
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
"""3차원 함수 시각화하기"""
def f(x, y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
```

```python
# Select data for each target class
data_class0 = iris.data[iris.target == 0]
data_class1 = iris.data[iris.target == 1]
data_class2 = iris.data[iris.target == 2]

# Plot histograms for each class
plt.hist(data_class0[:, 0], bins=20, alpha=0.5, label='Class 0')  # Using the first feature for class 0
plt.hist(data_class1[:, 0], bins=20, alpha=0.5, label='Class 1')  # Using the first feature for class 1
plt.hist(data_class2[:, 0], bins=20, alpha=0.5, label='Class 2')  # Using the first feature for class 2
```

```python
"""히스토그램 비교할 때는 histtype='stepfilled'와 몇몇 투명도 alpha를 결합하는 것이 매우 유용하다."""
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)

kwargs=dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
"""손으로 쓴 숫자"""
from sklearn.datasets import load_digits
digits=load_digits(n_class=6)

fig, ax= plt.subplots(8,8,figsize=(6,6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
"""2차원에 사영"""
from sklearn.manifold import Isomap
iso=Isomap(n_components=2)
projection=iso.fit_transform(digits.data)
```

```python
plt.scatter(projection[:,0], projection[:,1], lw=0.1, c=digits.target, cmap=plt.get_cmap('cubehelix',6))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5,5.5)
```
