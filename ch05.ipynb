<h2>ch5 머신러닝</h2>

머신러닝은 지도학습(supervised감독) 과 비지도 학습(unsupervised)의 두가지 주요 유형

1)지도학습
feature  입력데이터 
label    긍정/부정

지도학습의 모델을 이용해 알려지지 않은 새 데이터에 레이블을 적용할 수 있다.이것은 다시 분류와 회귀로 나뉜다.
2)비지도 학습
feature 모델링, 데이터세트가 스스로 말하게 하는 것
군집화, 차원축소와 같은 작업

<h3>지도학습-분류</h3>

y=ax+b가 우리가 찾으려는 규칙(모델)이라고 하자.
학습(러닝)을 통해 a,b(모델모수)를 찾아햐 한다. a, b를 가중치라고 한다.

3)예측
모델이 학습됐으니(가중치 a,b를 찾았으므로 y=ax+b 규칙이 완성된다.) 이제 그것을 레이블이 없는 새 데이터에 일반화 할 수 있다.
즉, 새로운 데이터 집합을 가져와 그 집합에 이 모델의 선을 그리고, 이 모델을 기반으로 새 점에 레이블을 할당할 수 있다.

Some important classification algorithms that we will discuss in more detail are Gaussian naive Bayes (see [In Depth: Naive Bayes Classification](05.05-Naive-Bayes.ipynb)), support vector machines (see [In-Depth: Support Vector Machines](05.07-Support-Vector-Machines.ipynb)), and random forest classification (see [In-Depth: Decision Trees and Random Forests](05.08-Random-Forests.ipynb)).

<h3>지도학습-회귀</h3>

레이블을 세번쨰 공간 차원(높이축)으로 취급했을 때, hyperplane(2차원이 아닌 다차원에서의 평면)을 데이터에 적합시킬 수 있다고 가정한다.
3차원 데이터를 관통하는 평면에 적합(fit)시키면 어떤 입력 매개변수에 대해서도 기대레이블을 예측할 수 있다는 말이 합리적으로 들린다.

<h3>비지도학습-군집화:레이블 없는 데이터에 대한 레이블 추론</h3>

 
군집화를 위한 예제 데이터


k-평균 군집화 모델로 레이블을 할당한 데이터

k-평균은 k개의 군집중심으로 구성된 모델에 적합(fit)시킨다.

[머신러닝] K-평균(K-Means) 알고리즘
https://velog.io/@jhlee508/%EB%A8%B8%EC%8B%A0%EB%9F%AC%EB%8B%9D-K-%ED%8F%89%EA%B7%A0K-Means-%EC%95%8C%EA%B3%A0%EB%A6%AC%EC%A6%98


K-means 알고리즘은 K-NN 알고리즘과 유사하면서도 다르다. 둘은 모두 K개의 점을 지정하여 거리를 기반으로 구현되는 거리기반 분석 알고리즘이다. 하지만 둘의 차이점은 지도학습에 속하는 K-NN 알고리즘과 달리 K-means 알고리즘은 비지도학습 방법에 속한다. 따라서 K-means 알고리즘의 목적을 분류(Classification)라고 하지 않고 군집화(Clustering)라고 한다. 즉, K-NN은 분류 알고리즘, K-means은 군집화 알고리즘이라고 할 수 있다.

We will discuss the *k*-means algorithm in more depth in [In Depth: K-Means Clustering](05.11-K-Means.ipynb).
Other important clustering algorithms include Gaussian mixture models (See [In Depth: Gaussian Mixture Models](05.12-Gaussian-Mixtures.ipynb)) and spectral clustering (See [Scikit-Learn's clustering documentation](http://scikit-learn.org/stable/modules/clustering.html)).

<h2>Estimator API</h2>

1. Scikit-Learn 으로 적절한 Estimator 클래스를 임포트해서 모델의 클래스를 선택한다. <br />

<h2>지도학습예제-간단한 선형회귀</h2>

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model
```

```python
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X, y)
```

```python
"""기울기"""
model.coef_
```

```python
"""절편"""
model.intercept_
```

```python
"""x값이 주어질 때, y값 예측"""
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

<h2>지도학습예제-붓꽃 분류-가우스 나이브 베이즈 모델 이용</h2>

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

<h2>비지도학습예제-붓꽃 차원-빠른 선형 차원 축소 기법인 주성분분석 사용<h2>

붓꽃 데이터의 feature이 4개였다=>붓꽃데이터는 4차원이다.
비지도학습은 특징배열X만 가지고 한다.

주성분분석 (PCA)를 통해 두 개의 성분, 즉 그 데이터의 2차원 표현을 반환하도록 요청한다.

```python
from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(X_iris)         # 4. Transform the data to two dimensions
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False);
```

<h2> 비지도학습:차원축소</h2>

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
            edgecolor='none', alpha=0.5,
          )
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5);
```
